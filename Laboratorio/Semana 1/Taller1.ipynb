## Fibonacci Problem

```python
#Paquetes a Utilizar
import matplotlib.pylab as plt
import time

#Fibonacci Recursivo
def fib_recursion(n):
    if(n == 0 or n == 1):
        n_fibonacci = n
    else:
        n_fibonacci = fib_recursion(n-1) + fib_recursion(n-2)
    return n_fibonacci

#Fibonacci Normal
def fib_normal(n):
    a = 1
    b = 0
    for i in range(n):
        a,b = a+b,a
    return b

#DefiniciÃ³n Variables y Listas para Plotear
list_of_nR = []
list_of_timesR = []
list_of_nN = []
list_of_timesN = []
convR = []
convN = []
nR = 30
nN = 100

#Encontrar Tiempo Gastado y Convergencia Fibonacci Recursivo
for i in range(1,nR):
    initR=time.time()
    fib_recursion(i)
    difference_timeR = time.time()-initR
    list_of_nR.append(i)
    list_of_timesR.append(difference_timeR)
    convR.append(fib_recursion(i+1)/fib_recursion(i))
    
#Encontrar Tiempo Gastado y Convergencia Fibonacci Normal
for i in range(1,nN):
    initN=time.time()
    fib_normal(i)
    difference_timeN = time.time()-initN
    list_of_nN.append(i)
    list_of_timesN.append(difference_timeN)
    convN.append(fib_normal(i+1)/fib_normal(i))

#GrÃ¡ficas
fig0 = plt.figure(figsize=(18,6))
ax0 = fig0.add_subplot(121)
ax0.plot(list_of_nR, list_of_timesR, '-r.', label='Recursive Fibonacci')
ax0.set_xlabel('$n$', size=12)
ax0.set_ylabel('Time Spent by the Function [$s$]', size=12)
ax0.set_title('Time Spent by Recursive Fibonacci', size=12)
ax0.legend()

ax1 = fig0.add_subplot(122)
ax1.plot(list_of_nN, list_of_timesN, '-k.', label='Normal Fibonacci')
ax1.set_xlabel('$n$', size=12)
ax1.set_ylabel('Time Spent by the Function [$s$]', size=12)
ax1.set_title('Time Spent by Normal Fibonacci', size=12)
ax1.legend()

plt.show()

fig1 = plt.figure(figsize=(18,6))
ax0 = fig1.add_subplot(121)
ax0.plot(list_of_nR, convR, '-r.', label='Recursive Fibonacci')
ax0.set_xlabel('$n$', size=12)
ax0.set_ylabel('Recursive Fibonacci(n+1) / Recursive Fibonacci(n)', size=12)
ax0.set_title('Convergence of Recursive Fibonacci', size=12)
ax0.legend()

ax1 = fig1.add_subplot(122)
ax1.plot(list_of_nN, convN, '-k.', label='Normal Fibonacci')
ax1.set_xlabel('$n$', size=12)
ax1.set_ylabel('Normal Fibonacci(n+1) / Normal Fibonacci(n)', size=12)
ax1.set_title('Convergence of Normal Fibonacci', size=12)
ax1.legend()

plt.show()
```

Al observar los tiempos de respuesta de las funciones Fibonacci Recursivo y Normal encontramos que hay varios ordenes de magnitud de diferencia, tal que la funciÃ³n recursiva para valores de n cercanos a 30 tarda al rededor de 0.2 segundos en responder, lo cual puede que no parezca mucho pero al compararlo con los tiempos del normal que para valores de n al rededor de 100 (con los cuales el recursivo empieza a tardar demasiado) tarda Ãºnicamente 4.5e-6 segundos, un tiempo bastante menor para un valor bastante mayor, lo cual hace que el mÃ©todo normal sea mucho mÃ¡s efectivo. Esto sucede debido a que el mÃ©todo recursivo requiere llamarse a si mismo 2 veces cada vez que halla un nÃºmero, lo que provoca un nÃºmero de iteraciones y procesos exponencialmente creciente y muy superior al nÃºmero de iteraciones relizadas por el mÃ©todo normal que puede hallar el valor de la suceciÃ³n de fibonacci de un n = 100000 con relativa rapidez. Las ventajas del mÃ©todos recursivo radican en que su implementaciÃ³n en cÃ³digo es mucho mÃ¡s intuitiva, y va mÃ¡s acorde a la definiciÃ³n comÃºn de la sucesiÃ³n de Fibonacci, la normal por su lado requiere algo mÃ¡s de entendimiento del funcionamiento de la misma.
