```python
def normalized(dat):
    N = len(dat.keys())
    for i in range (N):
        dat[dat.keys()[i]] = (dat[dat.keys()[i]] - np.mean(dat[dat.keys()[i]]))
    return dat
```

```python
dataNorm = normalized(data)
```

```python
def cova(dat):
    N = len(dat.keys())
    
    for i in range (N):
        dat[dat.keys()[i]] = (dat[dat.keys()[i]].T*dat[dat.keys()[i]])/N-1
    
    return dat
```

```python
dataCov = cova(dataNorm)
```
